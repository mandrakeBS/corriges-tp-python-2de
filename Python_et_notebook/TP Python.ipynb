# Détermination par balayage d'un encadrement de $\sqrt{2}$ d'amplitude $d\leqslant 10^{-n}$

```python

import numpy as np

def f(x):
    # Définissez ici votre fonction
    return x**2 - 2  # Par exemple, on cherche sqrt(2)

def find_bounds(n, x0, delta=0.1):
    epsilon = 10**(-n)
    a = x0  # Point de départ
    b = x0 + delta  # Point suivant
    fa = f(a)
    fb = f(b)

    # On cherche les bornes
    while abs(b - a) > epsilon:
        if fa * fb <= 0:  # Il y a un changement de signe
            break
        a = b
        b += delta
        fa = fb
        fb = f(b)

    return a, b

def main(n, x0):
    lower_bound, upper_bound = find_bounds(n, x0)
    print(f"Encadrement de la solution : [{lower_bound}; {upper_bound}]")
    print(f"Amplitude : {upper_bound - lower_bound}")

if __name__ == "__main__":
    n = 3  # Par exemple, 10^{-3}
    x0 = 0  # Point de départ pour le balayage
    main(n, x0)
```

# Détermination par balayage d'un encadrement de $\sqrt{2}$ d'amplitude $d\leqslant 10^{-n}$ (TP A1_1).

```python
# TP_A1_1

def f(x):
    # Définissez ici votre fonction
    return x**2 - 2  # Par exemple, on cherche sqrt(2)

def find_bounds(n, x0, delta=0.1):
    epsilon = 10 ** (-n)
    a = x0  # Point de départ
    b = x0 + delta  # Point suivant
    fa = f(a)
    fb = f(b)

    # Cherche l'encadrement
    while abs(b - a) > epsilon:
        if fa * fb > 0:  # Pas de changement de signe
            a = b
            b += delta
            fa = fb
            fb = f(b)
        else:  # Changement de signe trouvé
            return a, b

    return a, b

# Exécution du programme
#n = 3  # Par exemple, 10^{-3}
#x0 = 0  # Point de départ pour le balayage
lower_bound, upper_bound = find_bounds(3, 0, 0.1)
print(f"Encadrement de la solution : [{lower_bound:.{n}f}; {upper_bound:.{n}f}]")
print(f"Amplitude : {upper_bound - lower_bound:.{n}f}")
```

## Par dichotomie (TP A1_2).

```python
# TP_A1_2

def f(x):
    # Définissez ici votre fonction
    return x**2 - 2  # Par exemple, on cherche sqrt(2)

def dichotomy(n, a, b):
    epsilon = 10 ** (-n)
    
    # Vérification que f(a) et f(b) ont des signes opposés
    while abs(b - a) > epsilon:
        mid = (a + b) / 2
        if f(mid) == 0:  # Si on trouve une racine exacte
            return mid, mid
        elif f(a) * f(mid) < 0:  # Changement de signe entre a et mid
            b = mid
        else:  # Changement de signe entre mid et b
            a = mid

    return a, b

# Exécution du programme
n = 3  # Par exemple, 10^{-3}
a = 0  # Intervalle de départ
b = 2  # Intervalle de départ
lower_bound, upper_bound = dichotomy(n, a, b)
print(f"Encadrement de la solution : [{lower_bound:.{n}f}; {upper_bound:.{n}f}]")
print(f"Amplitude : {upper_bound - lower_bound:.{n}f}")
```

# Déterminer la première puissance d'un nombre positif donné supérieure ou inférieure à une valeur donnée (TP A3_1).

```python
# TP_A3_1
def first_power_greater_than(base, limit):
    """Détermine la première puissance de 'base' qui est supérieure à 'limit'."""
    exponent = 0
    while base ** exponent <= limit:
        exponent += 1
    return exponent

def first_power_less_than(base, limit):
    """Détermine la première puissance de 'base' qui est inférieure à 'limit'."""
    exponent = 0
    while base ** exponent >= limit:
        exponent -= 1
    return exponent

# Exécution du programme
base = 2  # Exemple de base (nombre positif)
limit_greater = 10  # Valeur limite pour la puissance supérieure
limit_less = 20  # Valeur limite pour la puissance inférieure

# Déterminer la première puissance supérieure
exponent_greater = first_power_greater_than(base, limit_greater)
print(f"La première puissance de {base} supérieure à {limit_greater} est : {base}^{exponent_greater} = {base ** exponent_greater}")

# Déterminer la première puissance inférieure
exponent_less = first_power_less_than(base, limit_less)
print(f"La première puissance de {base} inférieure à {limit_less} est : {base}^{exponent_less} = {base ** exponent_less}")
```

# Etudier l'alignement de trois points dans le plan

## Méthode 1: Calculer l'aire du triangle formé par les trois points. Cette aire est nulle ssi les trois points sont alignés (TP B3_1).  

```python
# TP_B3_1

def are_points_aligned(point1, point2, point3):
    x1, y1 = point1
    x2, y2 = point2
    x3, y3 = point3
    
    # Calcul de l'aire du triangle
    area = 0.5 * abs(x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2))
    
    return area == 0

# Exécution du programme
point1 = (1, 2)  # Exemple de point A
point2 = (2, 3)  # Exemple de point B
point3 = (3, 4)  # Exemple de point C

if are_points_aligned(point1, point2, point3):
    print("Les points sont alignés.")
else:
    print("Les points ne sont pas alignés.")
```

## Méthode avec le déterminants (TP B3_2)

```python
#TP B3_2

def are_points_aligned(point1, point2, point3):
    x1, y1 = point1
    x2, y2 = point2
    x3, y3 = point3
    
    # Calcul du déterminant d'ordre 2
    determinant = (x2 - x1) * (y3 - y1) - (y2 - y1) * (x3 - x1)
    
    return determinant == 0

# Exécution du programme
point1 = (1, 2)  # Exemple de point A
point2 = (2, 3)  # Exemple de point B
point3 = (3, 4)  # Exemple de point C

if are_points_aligned(point1, point2, point3):
    print("Les points sont alignés.")
else:
    print("Les points ne sont pas alignés.")
```

# Déterminer une équation de droite passant par deux points donnés $A(x_A;\,y_A)$ et $B(x_B;\,y_B)$.

## Avec la «formule» de la pente et l'équation réduite de la droite: $y=mx+p$ (TP_B3_2).

```python
# TP_B3_2

def equation_de_droite(point1, point2):
    x1, y1 = point1
    x2, y2 = point2
    
    # Calcul de la pente (m)
    if x2 == x1:  # Éviter la division par zéro
        return f"x = {x1}"  # Droite verticale

    m = (y2 - y1) / (x2 - x1)
    # Calcul de l'ordonnée à l'origine (p)
    p = y1 - m * x1
    
    # Forme explicite
    return f"y = {m:.2f}x + {p:.2f}"

# Exécution du programme
point1 = (1, 2)  # Exemple de point A
point2 = (3, 4)  # Exemple de point B

equation = equation_de_droite(point1, point2)
print(f"L'équation de la droite passant par les points {point1} et {point2} est : {equation}")
```

# Approximation numérique d'un extremum d'une fonction dont on connait le tableau de variation.

## Méthode par balayage.  
1) On définit la fonction dont on souhaite trouver les extremum.   
2) On détermine l'intervalle sur lequel on souhaite faire notre balayage.
3) On évalue la fonction sur des points de l'intervalle balayé et on identifie les valeurs maximale et minimale.

## Avec tuple, donc hors programme seconde.

```python

def f(x):
    # Définissez ici la fonction dont vous voulez trouver l'extrême
    return -1 * (x ** 2) + 4 * x  # Exemple : une parabole

def find_extremum(start, end, step):
    current_x = start
    max_value = float('-inf')
    min_value = float('inf')
    max_x = start
    min_x = start

    while current_x <= end:
        current_value = f(current_x)
        
        # Vérification du maximum
        if current_value > max_value:
            max_value = current_value
            max_x = current_x
        
        # Vérification du minimum
        if current_value < min_value:
            min_value = current_value
            min_x = current_x
            
        current_x += step

    return (max_x, max_value), (min_x, min_value)

# Exécution du programme
start = 0  # Borne inférieure
end = 5    # Borne supérieure
step = 0.1 # Pas de balayage

(max_point, max_value), (min_point, min_value) = find_extremum(start, end, step)

print(f"Maximum : f({max_point[0]}) = {max_value}")
print(f"Minimum : f({min_point[0]}) = {min_value}")
```

## Méthode par dichotomie.  
1) On définit la fonction dont on souhaite trouver les extremum.   
2) On détermine l'intervalle sur lequel on sait qu'il y a un extremum.
3) On évalue la fonction sur des points de l'intervalle balayé et on identifie les valeurs maximale et minimale. 

### Méthode des tiers

```python

def f(x):
    # Définissez ici la fonction dont vous voulez trouver l'extrême
    return -1 * (x ** 2) + 4 * x  # Exemple : une parabole

def find_maximum_by_dichotomy(a, b, epsilon):
    while (b - a) > epsilon:
        mid1 = a + (b - a) / 3  # Premier point médian
        mid2 = b - (b - a) / 3  # Deuxième point médian
        
        # Évaluer la fonction aux points médians
        f_mid1 = f(mid1)
        f_mid2 = f(mid2)
        
        # Comparer les valeurs pour réduire l'intervalle
        if f_mid1 < f_mid2:
            a = mid1  # Le maximum est dans [mid1, b]
        else:
            b = mid2  # Le maximum est dans [a, mid2]

    # Calculer le point moyen final pour l'extremum
    maximum_x = (a + b) / 2
    maximum_value = f(maximum_x)

    # Afficher le résultat
    print(f"Maximum : f({maximum_x}) = {maximum_value}")

# Exécution du programme
a = 0  # Borne inférieure
b = 5  # Borne supérieure
epsilon = 0.01  # Précision souhaitée

find_maximum_by_dichotomy(a, b, epsilon)
```

### Méthode de la bissection (TP_C3_2).

```python
# TP_C3_2

def f(x):
    # Définissez ici la fonction dont vous voulez trouver l'extrême
    return -1 * (x ** 2) + 4 * x  # Exemple : une parabole

def find_maximum_by_dichotomy(a, b, epsilon):
    while (b - a) > epsilon:
        mid = (a + b) / 2  # Point médian
        
        # Évaluer la fonction aux points a, mid et b
        f_a = f(a)
        f_mid = f(mid)
        f_b = f(b)

        # Comparer les valeurs pour réduire l'intervalle
        if f_mid > f_a and f_mid > f_b:
            # Si mid est le maximum, nous réduisons l'intervalle à [a, b]
            a = mid
            # Ajustement pour garder b à une distance de epsilon
            b = (a + b) / 2
        elif f_a > f_mid and f_a > f_b:
            # Si a est le maximum, nous réduisons l'intervalle à [a, mid]
            b = mid
        else:
            # Si b est le maximum, nous réduisons l'intervalle à [mid, b]
            a = mid

    # Calculer le point moyen final pour l'extremum
    maximum_x = (a + b) / 2
    maximum_value = f(maximum_x)

    # Afficher le résultat
    print(f"Maximum : f({maximum_x}) = {maximum_value}")

# Exécution du programme
a = 0  # Borne inférieure
b = 5  # Borne supérieure
epsilon = 0.01  # Précision souhaitée

find_maximum_by_dichotomy(a, b, epsilon)
```
